```python
import re
import string
from Porter_Stemmer import PorterStemmer
import pandas as pd

sentence_file = open('Project4_sentences(2)/sentences.txt').readlines()
stop_words = open('Project4_sentences(2)/stop_words.txt').read().split()
p = PorterStemmer()
```

```python
def process_word(word):
    word = (''.join([x for x in word if x.isalpha()])).lower()
    if word in stop_words:
        return ''
    else: 
        return p.stem(word,0,len(word)-1)
```

```python
processed_sentence_list = []
for sentence in sentence_file:
    sentence_list = sentence.split()
    processed_sentence = [process_word(word) for word in sentence_list]
    #get rid of empty strings
    processed_sentence = ' '.join(processed_sentence).split()
    processed_sentence_list.append(processed_sentence)
```

```python
processed_sentence_list = []
for sentence in sentence_file:
    sentence_list = sentence.split()
    processed_sentence = [process_word(word) for word in sentence_list]
    #get rid of empty strings
    processed_sentence = ' '.join(processed_sentence).split()
    processed_sentence_list.append(processed_sentence)

words_list = []
for sentences in processed_sentence_list:
    words_list += sentences
words_series = pd.Series(words_list)
frequent_words_series = words_series.value_counts()[words_series.value_counts() > 2]

frequent_words = [word for word,value in frequent_words_series.iteritems()]
```

```python
tdm = pd.DataFrame(columns = frequent_words)

for sentence in processed_sentence_list:
    sentence_counts = {w:sentence.count(w) for w in frequent_words}
    tdm = tdm.append(sentence_counts,ignore_index = True)
```

```python
def distance(x, y):
    return sum([(y_val-x_val)**2 for x_val, y_val in zip(x,y)])**(1/2)
def fcan(df, learning_rate, max_radius):
    clusters = []
    clusters.append([df.iloc[0].values,1])
    for i in range(1, len(df)):
        current_pattern = df.iloc[i].values
#         print('----------------------------------------------')

#         print('Before iteration', i, 'clusters are', clusters)
        min_distance_cluster = pd.Series({'Index':-1,'Distance':100000000})

        for c in range(len(clusters)):
#             print(current_pattern,clusters[c][0])
            dist_to_cluster = distance(current_pattern,clusters[c][0])
#             print('Distance between',clusters[c][0], 'and', current_pattern,': ',dist_to_cluster)
            if dist_to_cluster <= max_radius and dist_to_cluster < min_distance_cluster.Distance:    
#                 print('Cluster for point', current_pattern, 'to join:', clusters[c][0])
                min_distance_cluster.Index = c
                min_distance_cluster.Distance = dist_to_cluster
        
        if min_distance_cluster.Index == -1:
            clusters.append([current_pattern, 1])
#             print('Creating new cluster at', current_pattern,'\n')
        else:
            clusters[min_distance_cluster.Index][0] = (clusters[min_distance_cluster.Index][1] * clusters[min_distance_cluster.Index][0] + learning_rate * current_pattern)/(clusters[min_distance_cluster.Index][1] + 1)
            clusters[min_distance_cluster.Index][1] += 1
    for cluster in clusters:
        print('Cluster:', list(cluster[0]))
        print('Number of Points in Cluster:', cluster[1])
        print('------------------------------------------')
# fcan(tdm, 1, 3)
```
