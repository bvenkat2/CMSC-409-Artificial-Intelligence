# Code

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_a = pd.read_csv('train_data_1.txt', header = None)
df_a.columns = ['Time','Energy']
```

```python
df_b = pd.read_csv('train_data_2.txt', header = None)
df_b.columns = ['Time','Energy']
```

```python
df_c = pd.read_csv('train_data_3.txt', header = None)
df_c.columns = ['Time','Energy']
```

```python
test = pd.read_csv('test_data_4.txt', header = None)
test.columns = ['Time','Energy']
```

```python
def normalize(df):
    ymin = df['Energy'].min()
    ymax = df['Energy'].max()
    xmin = df['Time'].min()
    xmax = df['Time'].max()
    df['Energy']=df['Energy'].apply(lambda y: (y-ymin)/(ymax-ymin))
    df['Time']=df['Time'].apply(lambda x: (x-xmin)/(xmax-xmin))
```

```python
normalize(df_a)
normalize(df_b)
normalize(df_c)
normalize(test)
```

```python
plt.scatter(df_a['Time'], df_a['Energy'], color = 'Red')
plt.scatter(df_b['Time'], df_b['Energy'], color = 'Green')
plt.scatter(df_c['Time'], df_c['Energy'], color = 'Blue')
```

```python
def update_weights(weights, learning_rate, output, desired, x_value):
    x_vector = [(x_value**(power)) for power in range(len(weights)-1, -1, -1)]
    increment = [2*learning_rate * (desired-output) * x for x in x_vector]
    updated = [x+y for x,y in zip(weights, increment)]
    return updated
    
```

```python
import math
def linear_activation(weights, x):
    return sum([(x**(power))*weight for weight,power in zip(weights, range(len(weights)-1, -1, -1))])    
```

```python
import random
import sys
def train_perceptron(learning_rate, df, max_iterations,poly, weights = None):
    if weights  == None:
        weights = [random.uniform(-.5,.5) for x in range(poly + 1)]
    print('Initial Weights: ', weights)
    for x in range(max_iterations):
        error = 0
        for i in range(len(df)):
            current_point = df.iloc[i]
            output = linear_activation(weights, current_point['Time'])
            weights = update_weights(weights, learning_rate, output, current_point['Energy'],current_point['Time'])
            error += (current_point['Energy']-output)**2
    print('Final Weights: ', weights)
    return weights
```

```python
def plotGraph(df,weights, group):
    x = np.linspace(0, 1, 100)
    y = sum([(x**(power))*weight for weight,power in zip(weights, range(len(weights)-1, -1, -1))])
    plt.plot(x, y)
    plt.scatter(df['Time'],df['Energy'])
    plt.title('Energy vs. Time ' +group)
    plt.xlabel('Time')
    plt.ylabel('Energy Consumption')
```

```python
def getError(df, weights):
    error = 0
    for i in range(len(df)):
        output = linear_activation(weights, df.iloc[i]['Time'])
        desired = df.iloc[i]['Energy']
        error += (desired-output)**2
    return error
```

# Outputs

## Architecture A (Linear Regression)

### Training Set (Day 1)

```python
weights= train_perceptron(.1, df_a, 1500, 1)
plotGraph(df_a,weights, "Day 1 (Training)" )
print('Total Training Error: ' , getError(df_a, weights))
```

### Training Set (Day 2)

```python
weights= train_perceptron(.1, df_b, 1500, 1, weights)
plotGraph(df_b,weights, "Day 2 (Training)" )
print('Total Training Error: ' , getError(df_b, weights))
```

### Training Set (Day 3)

```python
weights= train_perceptron(.1, df_c, 1500, 1, weights)
plotGraph(df_c,weights, "Day 3 (Training)" )
print('Total Training Error: ' , getError(df_c, weights))
```

### Testing Set (Day 4)

```python
plotGraph(test, weights,"Day 4 (Testing)" )
print('Total Testing Error: ', getError(test, weights))
```
